```python
import plotly.express as px
import geopandas as gpd
```

```python
dados = pd.read_csv('dados/hotosm_tur_destroyed_buildings_polygons_csv.csv')
dados.head()
```

```python
fig = px.scatter_mapbox(
    dados,
    lat=dados['latitude'],
    lon=dados['longitude'],
    zoom=7,
    height=300,
    color_discrete_sequence=['red']
)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data=go.Scattergeo(
        lon = dados['longitude'],
        lat = dados['latitude'],
        #text = df['text'],
        mode = 'markers',
        #marker_color = df['cnt'],
        ))

fig.update_layout(
        title = 'Edificio destruidos pelo o terremoto na Turquia',
        mapbox_style="basic",
        mapbox_zoom=9
    )
fig.update_geos(
    projection_type='robinson',
    landcolor='white',
    oceancolor='blue',
    showcountries=True,
    countrycolor='black',
    lonaxis_range=[25, 50],
    lataxis_range=[35, 43],
    
)
fig.show()
```
